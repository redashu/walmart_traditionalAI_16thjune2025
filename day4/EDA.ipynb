```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("walmart_sales.csv", parse_dates=["date"])

# Display first few rows
print(df.head())
```

```python
# Basic structure
print(df.info())

# Summary statistics
print(df.describe())
```

```python

print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['daily_sales'], kde=True, bins=30)
plt.title("Distribution of Daily Sales")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x='date', y='daily_sales', data=df)
plt.title("Walmart Store Daily Sales Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['daily_sales'])
plt.title("Boxplot of Daily Sales")
plt.show()
```

Optional: Add Derived Columns (Feature Engineering for EDA)

```python
df['day_of_week'] = df['date'].dt.day_name()
df['month'] = df['date'].dt.month

# Sales by weekday
plt.figure(figsize=(8, 4))
sns.boxplot(x='day_of_week', y='daily_sales', data=df, order=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title("Sales by Day of the Week")
plt.xticks(rotation=45)
plt.show()
```
